# First Test: Classify simulated data

## The variant parameter is the noise level of the data
## The Cole-Cole parameters are exact for each classification
## Evaluate effectiveness of EACH model

```python
#Create classifiers for simulated data

#Load simulated data of varying noise levels
classes = "18OctoberSim/simClasses.csv"
data01 = "18OctoberSim/simLowData.csv"
data10 = "18OctoberSim/simHighData.csv"
#...
titles = ["1% noise","10% noise"]

n_dt = 2 #Number of data sets (noise steps)
n_ex = 3 #Number of extractors
n_cl = 3 #Number of classifiers

SimModels = []
for i in c.All:
    for j in e.All:
        SimModels.append(dmlt.Model(j,i,order,zeroth))

#For each model, load a seperate Token for each data set
SimData = []
for model in SimModels:
    SimData.append(model.load(classes,data01))
    SimData.append(model.load(classes,data10))
    #...
```

```python
#Generate variances
varReal = []
varImag = []
zeroth = 0
for model in SimData:
    varReal.append(model.getVariances(0,0,zeroth))
    varImag.append(model.getVariances(1,0,zeroth))

#Replace each array with a set of tuples (index,value) of the n highest variances
for model in range(len(varReal)):
    tmp1 = []
    tmp2 = []
    for i in range(n):
        tmp1.append(varReal[model].index(max(varReal[model])))
        varReal[model][tmp1[-1]] = -1

        tmp2.append(varImag[model].index(max(varImag[model])))
        varImag[model][tmp2[-1]] = -1
    varReal[model] = tmp1
    varImag[model] = tmp2

#Order things nicely
var = []
for i in range(len(varReal)):
    #SimModels[i/2].train(SimData[i],[])
    if (i/6)%2 == 0:
        #var.append((SimModels[i/2].C.name,SimModels[i/2].L.name,"r"))
        var.append(varReal[i])
    else:
        #var.append((SimModels[i/2].C.name,SimModels[i/2].L.name,"i"))
        var.append(varImag[i])
```

```python
#Visualize data

#1%

green_patch = mpatches.Patch(color='green', label='Healthy')
blue_patch = mpatches.Patch(color='blue', label='Benign')
red_patch = mpatches.Patch(color='red', label='Malignant')

plt.legend(handles=[green_patch,blue_patch,red_patch])
imaginary = 0
feats = varReal[0]
plt.xscale('log')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Capacitance [F]')
#plt.yscale('log')
for i in range(len(SimData[0].getAllY())):
    if SimData[0].getAllY()[i] == positive:
        SimModels[0].plotSample(SimData[0],i,'g--',imaginary)
    elif SimData[0].getAllY()[i] == "Benign":
        SimModels[0].plotSample(SimData[0],i,'b--',imaginary)
    elif SimData[0].getAllY()[i] == "Malignant":
        SimModels[0].plotSample(SimData[0],i,'r--',imaginary)
plt.show()

plt.legend(handles=[green_patch,blue_patch,red_patch])
imaginary = 1
feats = varImag[0]
plt.xscale('log')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Conductance [S]')
#plt.yscale('log')
for i in range(len(SimData[0].getAllY())):
    if SimData[0].getAllY()[i] == positive:
        SimModels[0].plotSample(SimData[0],i,'g--',imaginary)
    elif SimData[0].getAllY()[i] == "Benign":
        SimModels[0].plotSample(SimData[0],i,'b--',imaginary)
    elif SimData[0].getAllY()[i] == "Malignant":
        SimModels[0].plotSample(SimData[0],i,'r--',imaginary)
plt.show()

#10%

plt.legend(handles=[green_patch,blue_patch,red_patch])
imaginary = 0
feats = varReal[0]
plt.xscale('log')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Capacitance [F]')
#plt.yscale('log')
for i in range(len(SimData[1].getAllY())):
    if SimData[1].getAllY()[i] == positive:
        SimModels[1].plotSample(SimData[1],i,'g--',imaginary)
    elif SimData[1].getAllY()[i] == "Benign":
        SimModels[1].plotSample(SimData[1],i,'b--',imaginary)
    elif SimData[1].getAllY()[i] == "Malignant":
        SimModels[1].plotSample(SimData[1],i,'r--',imaginary)
plt.show()

plt.legend(handles=[green_patch,blue_patch,red_patch])
imaginary = 1
feats = varImag[0]
plt.xscale('log')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Conductance [S]')
#plt.yscale('log')
for i in range(len(SimData[1].getAllY())):
    if SimData[1].getAllY()[i] == positive:
        SimModels[1].plotSample(SimData[1],i,'g--',imaginary)
    elif SimData[1].getAllY()[i] == "Benign":
        SimModels[1].plotSample(SimData[1],i,'b--',imaginary)
    elif SimData[1].getAllY()[i] == "Malignant":
        SimModels[1].plotSample(SimData[1],i,'r--',imaginary)
plt.show()
```

```python
## Calculate average relative error:
#Sum of the squares of the differences between each point and the approximation
#Divided by the standard deviation of all points for normalization
#Averaged across each sweep

print "\t\tHealthy\t\tCancer"
for model in range(n_ex)[:1]:
    sumHealthy = 0.0
    numHealthy = 0.0
    sumCancer = 0.0
    numCancer = 0.0
    for row in range(CancerData[model].size):
        #Real
        feats = varReal[model]
        imaginary = 0
        realPart = CancerModels[model].plotApproximation(CancerData[model],row,'',imaginary,0,feats)
        #plt.plot(realPart)

        #Imaginary
        feats = varImag[model]
        imaginary = 1
        imagPart = CancerModels[model].plotApproximation(CancerData[model],row,'',imaginary,0,feats)
        
        approx = np.absolute(np.asarray(realPart) + 1j*np.asarray(imagPart))[:400]
        actual = np.absolute(np.asarray(CancerData[model].getAllData()[0][row]))[:400]

        if CancerData[model].getAllY()[row] == positive:
            sumHealthy += sum(np.abs(approx-actual))/np.std(np.abs(approx-actual))
            numHealthy += 1
        else:
            sumCancer += sum(np.abs(approx-actual))/np.std(np.abs(approx-actual))
            numCancer += 1
    print "%s\t%f\t%f"%(CancerModels[model].C.name,sumHealthy/numHealthy,sumCancer/numCancer)
```
